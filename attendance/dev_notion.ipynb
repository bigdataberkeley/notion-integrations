```python
from utils import *
from collections import defaultdict

from keys import EVENTS_DATABASE_ID, ATTENDANCE_TRACKER_DATABASE_ID
```

```python
for user in get_all_users():
    if user['name'] not in ignored_users:
        person_id_to_user[user['id']] = user
```

```python
for p in get_pages(EVENTS_DATABASE_ID):
    event_id_to_event[p['id']] = p 
```

```python
attendance_sheets = get_pages(ATTENDANCE_TRACKER_DATABASE_ID)
for s in attendance_sheets:
    load_person(s)
```

```python
person_to_events = defaultdict(list)   # person_id -> list[(event_id, status)]

# Extract attendance information from all events
def sort_key(event):
    try:
        props = event['properties']
        date = props['Date']['date']
        start = date['start'][:10]
        time = int(datetime.datetime.strptime(start, "%Y-%m-%d").timestamp())
        return time
    except Exception as e:
        print(event)
        raise e

events = list(event_id_to_event.values())
events.sort(key=sort_key)    # Parse events in calendar order so attendance log is in calendar order
for event in events:
    props = event["properties"]
    title = props['Name']['title'][0]['plain_text']
    event_id = event['id']
    date = props['Date']['date']['start']

    print(f'Extracting attendance for event {title} on {date}')
    statuses = ['Attended', 'Late', 'Approved Absence', 'Excused Absence', 'Unexcused Absence']
    for status in statuses:
        people = props[status]['people']
        for person in people:
            person_id = person['id']
            if person_id in person_id_to_user:
                person_to_events[person_id].append((event_id, status))
```

```python
# Added needed attendance sheets for all members:
for person_id in person_id_to_user:
    if person_id not in person_to_attendance_sheet:
        p = get_person(person_id)
        add_person_attendance(person_id)
        print(f'Added missing attendence sheet for {p['name']} with id {person_id}')
```

```python
for person_id, event_logs in person_to_events.items():
    update_attendance(person_id, event_logs) 
```
